# Financial Markets as a Random Walk

```r
library(data.table)
```

```r
## used each of these seeds in turn to produce the plots
## set.seed(1)
## set.seed(100)
## set.seed(30)

N <- 10000
x <- cumsum(sample(c(-1, 1), N, TRUE))
plot(x)
```

```r
cor(x, shift(x), use = "complete.obs")
```

```r
cor(diff(x), shift(diff(x)), use = "complete.obs")
```
